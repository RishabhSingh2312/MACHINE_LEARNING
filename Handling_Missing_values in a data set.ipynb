# Imputation missing values

## 1- Mean Value Imputation:

```python
df["Age_mean"]=df["age"].fillna(df["age"].mean())
```

```python
df[["Age_mean", "age"]]
```

```python
df["age"].isnull().sum()
```

```python
df["Age_mean"].isnull().sum()
```

```python
df.head()
```

```python
df.drop("age_mean", inplace= True,axis=1)
```

```python
df
```

```python
## Mean manupulation works well when we have nomally distributed data
```

# Median value imputation - if we have outliers in the data set

```python
df["Age_median"]=df["age"].fillna(df["age"].median())
```

```python
df
```

# Mode Imputation Technique_ "When we have categorical type of data set"

```python
df[df["embarked"].isnull()]
```

```python
df["embarked"].unique()
```

```python
mode_value=df[df["embarked"].notna()]["embarked"].mode()[0]
```

```python
mode_value
```

```python
df["Embarked_mode"]=df["embarked"].fillna(mode_value)
```

```python
df
```
